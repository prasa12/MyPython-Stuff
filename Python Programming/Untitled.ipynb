```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Read the image using OpenCV
I = cv2.imread('00-puppy.jpg')

# Convert the image from BGR to RGB (OpenCV reads images in BGR format)
I_rgb = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)

# Create a mask with ones
mask = np.ones(I_rgb.shape[:2], dtype=np.uint8)

# Set a specific region to zero
mask[10:20, 10:20] = 0
num_bins = 6
range_min = 0
range_max = 256


# Calculate histogram for each channel using the mask
# OpenCV's calcHist requires a list of images, so we pass [I_rgb]
hist_red = cv2.calcHist([I_rgb],[0],None,[num_bins],[0, 256])
#hist_green = cv2.calcHist([I_rgb], [1], mask, [256], 26, [0, 256])
#hist_blue = cv2.calcHist([I_rgb], [2], mask, [256], [0, 256])

# Plot the histograms
plt.figure(figsize=(12, 6))
plt.plot(hist_red, color='r', label='Red Channel')
bin_edges = np.linspace(range_min, range_max, num_bins + 1)
print("Bin edges:", bin_edges)
#plt.plot(hist_green, color='g', label='Green Channel')
#plt.plot(hist_blue, color='b', label='Blue Channel')
plt.title('Histogram of Masked Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Count')
plt.legend()
plt.show()
```
